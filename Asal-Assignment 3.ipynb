## Exercise 3: Manage Missing Data
We’ve got a number of ways in general of dealing with missing data. These involve

1. Dropping off cases (or rows) in the data with any missing variables
2. Excluding variables in the data with any missing data 
3. Selectively choosing indicators with only a limited amount of missing data
4. Replacing missing variables with averages, or other representative values
5. Creating a separate model to predict missing data

- Count the missing values in each column
- Manage the missing values (delete or replace values or leave them as they are) and briefly explain your choice for each column


```python
missing_values_count = df.isnull().sum()  
print(missing_values_count)
```

```python
# Replace missing values in 'rating' with the mean rating  
df['rating'] = df['rating'].fillna(df['rating'].mean()) 

# Replace missing values in 'review_count' with 0  
df['review_count'] = df['review_count'].fillna(0) 

# Drop rows with missing 'isbn'  
df = df.dropna(subset=['isbn'])  

# Replace missing values in 'booktype' with 'unknown'  
df['booktype'] = df['booktype'].fillna('unknown')  

# Replace missing author URLs with 'Not available'  
df['author_url'] = df['author_url'].fillna('Not available')  

# Replace missing values in 'year' with median  
df['year'] = df['year'].fillna(df['year'].median())  

# Replace missing values in 'genre_urls' with 'Not specified'  
df['genre_urls'] = df['genre_urls'].fillna('Not specified')  

# Drop rows with missing 'dir'  
df = df.dropna(subset=['dir'])  

# Replace missing values in 'rating_count' with 0  
df['rating_count'] = df['rating_count'].fillna(0)  

# Drop rows with missing 'name'  
df = df.dropna(subset=['name']) 
```

```python
1. Rating
Action: Replace any missing ratings with the average (mean) rating.
Reason: Ratings are numbers that usually follow a normal distribution. Using the average helps keep the overall average stable without significantly changing the data patterns.

2. Review Count
Action: Change missing review counts to 0.
Reason: If there are no reviews counted, it makes sense to show 0, indicating that the book hasn’t been reviewed yet.

3. ISBN
Action: Remove any rows that are missing ISBN numbers.
Reason: The ISBN is a unique identifier for books, and without it, you can’t accurately identify or reference the book.

4. Book Type
Action: Fill in missing book types with a default value like "unknown."
Reason: This allows you to keep all records in the dataset while noting that the specific book type isn't provided.

5. Author URL
Action: Either leave missing author URLs as they are or replace them with "Not available."
Reason: If the author's URL isn’t essential for your analysis, it’s okay not to fill it in or to mark it as unavailable.

6. Year
Action: Replace missing years with the median year.
Reason: The median is less impacted by extreme values compared to the average, which is helpful when dealing with books from different time periods.

7. Genre URLs
Action: Either keep the missing values or replace them with "Not specified."
Reason: If genre information isn’t crucial for your analysis, it’s fine to leave it missing or provide a placeholder.

8. Directory
Action: Remove any rows that are missing directory values.
Reason: If the directory is necessary to access the book's information, it’s better to delete those entries to make sure all remaining entries are valid.

9. Rating Count
Action: Fill in missing rating counts with 0.
Reason: Just like with the review count, a missing rating count can be seen as no ratings at all, so setting it to 0 clarifies the situation.

10. Name
Action: Remove any rows that have missing names.
Reason: The name of the book is essential for proper identification, so missing names should be cleaned from the dataset to maintain its integrity.
```
